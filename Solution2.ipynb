```python
from math import sqrt,pow
import matplotlib.pyplot as plt

y=[]
x=[]
def agm(x, y, tol=pow(10,10)):          
    xn, yn = (x + y) / 2.0, sqrt(x * y)
    while abs(xn - yn) > tol:   #looping  until abs(x-y)>1010
        xn, yn = (xn + yn) / 2.0, sqrt(xn * yn)
    return xn

 
def pairs(lower_lim,upper_lim):
    
    for i in range(lower_lim,upper_lim):    # Nested loop for all possible pairs 
        for j in range(lower_lim,upper_lim):
            if(i<j):
                AGM=agm(i,j)
                y.append(AGM)    #appending AGM for each distinct pair into list y
                x_axes="("+str(i)+","+str(j)+")" 
                x.append(x_axes) #Appending String of pairs in x[]Corresponding to AGM in y[]
  

  
lower_lim,upper_lim=[1,12]   #giving upper limit and lower limit manually
pairs(lower_lim,upper_lim) 
fig, ax = plt.subplots(figsize=(50, 10))
ax.plot(x,y)
plt.show()





```
